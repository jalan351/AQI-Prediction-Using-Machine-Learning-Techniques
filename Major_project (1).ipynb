CALCULATE AIR QUALITY INDEX FOR SO2 BASED ON FORMULA The air quality index is a piecewise linear function of the pollutant concentration. At the boundary between AQI categories, there is a discontinuous jump of one AQI unit. To convert from concentration to AQI this equation is used

```python
def cal_SOi(so2):
    si=0
    if (so2<=40):
     si= so2*(50/40)
    elif (so2>40 and so2<=80):
     si= 50+(so2-40)*(50/40)
    elif (so2>80 and so2<=380):
     si= 100+(so2-80)*(100/300)
    elif (so2>380 and so2<=800):
     si= 200+(so2-380)*(100/420)
    elif (so2>800 and so2<=1600):
     si= 300+(so2-800)*(100/800)
    elif (so2>1600):
     si= 400+(so2-1600)*(100/800)
    return si
df4['SOi']=df4['so2'].apply(cal_SOi)
data= df4[['so2','SOi']]
data.head()
# calculating the individual pollutant index for so2(sulphur dioxide)
```

```python
def cal_Coi(co):
    coi=0
    if(co<=40):
     coi= co*50/40
    elif(co>40 and co<=80):
     coi= 50+(co-40)*(50/40)
    elif(co>80 and co<=180):
     coi= 100+(co-80)*(100/100)
    elif(co>180 and co<=280):
     coi= 200+(co-180)*(100/100)
    elif(co>280 and co<=400):
     coi= 300+(co-280)*(100/120)
    else:
     coi= 400+(co-400)*(100/120)
    return coi
df4['Coi']=df4['co'].apply(cal_Coi)
data= df4[['co','Coi']]
data.head()
# calculating the individual pollutant index for co(carbon monoxide)
```

```python
def cal_Noi(no2):
    ni=0
    if(no2<=40):
     ni= no2*50/40
    elif(no2>40 and no2<=80):
     ni= 50+(no2-40)*(50/40)
    elif(no2>80 and no2<=180):
     ni= 100+(no2-80)*(100/100)
    elif(no2>180 and no2<=280):
     ni= 200+(no2-180)*(100/100)
    elif(no2>280 and no2<=400):
     ni= 300+(no2-280)*(100/120)
    else:
     ni= 400+(no2-400)*(100/120)
    return ni
df4['Noi']=df4['no2'].apply(cal_Noi)
data= df4[['no2','Noi']]
data.head()
# calculating the individual pollutant index for no2(nitrogen dioxide)
```

```python
def cal_O3i(o3):
    o3i=0
    if(o3<=40):
     o3i= o3*50/40
    elif(o3>40 and o3<=80):
     o3i= 50+(o3-40)*(50/40)
    elif(o3>80 and o3<=180):
     o3i= 100+(o3-80)*(100/100)
    elif(o3>180 and o3<=280):
     o3i= 200+(o3-180)*(100/100)
    elif(o3>280 and o3<=400):
     o3i= 300+(o3-280)*(100/120)
    else:
     o3i= 400+(o3-400)*(100/120)
    return o3i
df4['O3i']=df4['o3'].apply(cal_O3i)
data= df4[['o3','O3i']]
data.head()
# calculating the individual pollutant index for o3(ozone)
```

```python
def cal_Pmi(pm10):
    pmi=0
    if(pm10<=40):
     pmi= pm10*50/40
    elif(pm10>40 and pm10<=80):
     pmi= 50+(pm10-40)*(50/40)
    elif(pm10>80 and pm10<=180):
     pmi= 100+(pm10-80)*(100/100)
    elif(pm10>180 and pm10<=280):
     pmi= 200+(pm10-180)*(100/100)
    elif(pm10>280 and pm10<=400):
     pmi= 300+(pm10-280)*(100/120)
    else:
     pmi= 400+(pm10-400)*(100/120)
    return pmi
df4['Pmi']=df4['pm10'].apply(cal_Pmi)
data= df4[['pm10','Pmi']]
data.head()
# calculating the individual pollutant index for PM10
```

```python
def cal_PM25i(pm25):
    pM25i=0
    if(pm25<=40):
     pM25i= pm25*50/40
    elif(pm25>40 and pm25<=80):
     pM25i= 50+(pm25-40)*(50/40)
    elif(pm25>80 and pm25<=180):
     pM25i= 100+(pm25-80)*(100/100)
    elif(pm25>180 and pm25<=280):
     pM25i= 200+(pm25-180)*(100/100)
    elif(pm25>280 and pm25<=400):
     pM25i= 300+(pm25-280)*(100/120)
    else:
     pM25i= 400+(pm25-400)*(100/120)
    return pM25i
df4['PM25i']=df4['pm25'].apply(cal_PM25i)
data= df4[['pm25','PM25i']]
data.head()
# calculating the individual pollutant index for PM2.5
```

function to calculate the air quality index (AQI) of every data value

```python
def cal_aqi(si,coi,ni,pmi,pM25i, o3i):
    aqi=0
    if(si>coi and si>ni and si>pmi and si>pM25i and si> o3i):
     aqi=si
    if(coi>si and coi>ni and coi>pmi and coi>pM25i and coi> o3i):
     aqi=coi
    if(ni>si and ni>coi and ni>pmi and ni>pM25i and ni> o3i):
     aqi=ni
    if(pmi>si and pmi>coi and pmi>ni and pmi>pM25i and pmi>o3i):
     aqi=pmi
    if(pM25i>si and pM25i>coi and pM25i>ni and pM25i>pmi and pM25i> o3i):
     aqi=pM25i
    if(o3i>si and  o3i>coi and  o3i>ni and  o3i>pmi and o3i>pM25i):
     aqi= o3i


    return aqi

df4['AQI']=df4.apply(lambda x:cal_aqi(x['SOi'],x['Coi'],x['Noi'],x['O3i'],x['Pmi'],x['PM25i']),axis=1)
data= df4[['Unnamed: 0','SOi','Coi','Noi','O3i','Pmi','PM25i','AQI']]
data.head()
```

```python
data.head()
```

Splitting the data into dependent and independent column

```python
X=data[['SOi','Coi','Noi','O3i','Pmi','PM25i']]
Y=data['AQI']
X.head()
```

```python
Y.head()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=70)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
# splitting the data into training and testing data
```

```python
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
#predicting train
train_pred=model.predict(X_train)
#predicting on test
test_pred=model.predict(X_test)
```

```python
RMSE_train=(np.sqrt(metrics.mean_squared_error(Y_train,train_pred)))
RMSE_test=(np.sqrt(metrics.mean_squared_error(Y_test,test_pred)))
print("RMSE TrainingData = ",str(RMSE_train))
print("RMSE TestData = ",str(RMSE_test))
print('-'*50)
print('RSquared value on train:',model.score(X_train, Y_train))
print('RSquared value on test:',model.score(X_test, Y_test))
```
